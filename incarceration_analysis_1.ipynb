
# Data Analysis of Incarceration & Disenfranchisement

**Tools:** Python, Pandas, Matplotlib  
**Datasets used:**  
- `incarceration.csv` with incarceration rates for White and Black populations from **2011–2021**
- `disenfranchisement_data.csv` with disenfranchisement rates by race/ethnicity for **2022**

## Project Overview
This notebook analyzes racial disparities in incarceration and voting access using two small public-interest datasets. The goal is to identify patterns, visualize gaps, and connect the numbers to larger social issues.

## Research Questions
- How has the black incarceration rate per 100,000 population changed from 2005 to present compared to white incarceration rates
- How does the disenfranchisement for Black people compare to White people?
​


```python

import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path

pd.set_option('display.max_columns', 50)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')

base_path = Path('../data')
output_dir = Path('../outputs/charts')
output_dir.mkdir(parents=True, exist_ok=True)
```


## 1. Load the Data


```python

incarceration = pd.read_csv(base_path / 'incarceration.csv')
disenfranchisement = pd.read_csv(base_path / 'disenfranchisement_data.csv')

print("Incarceration data:")
display(incarceration)

print("Disenfranchisement data:")
display(disenfranchisement)
```


## 2. Clean and Prepare the Data
We standardize column names and convert percentages into numeric values.


```python

incarceration.columns = incarceration.columns.str.strip().str.lower()
disenfranchisement.columns = disenfranchisement.columns.str.strip().str.lower()

disenfranchisement['value_numeric'] = (
    disenfranchisement['value']
    .str.replace('%', '', regex=False)
    .astype(float)
)

incarceration.head(), disenfranchisement.head()
```


## 3. Quick Summary Statistics


```python

print("Incarceration summary:")
display(incarceration.describe())

print("Disenfranchisement rates:")
display(disenfranchisement[['category', 'value', 'value_numeric']])
```


## 4. Reshape the Incarceration Data
To make plotting easier, we convert the incarceration table from wide format to long format.


```python

incarceration_long = incarceration.melt(
    id_vars='year',
    value_vars=['white', 'black'],
    var_name='race',
    value_name='incarceration_rate'
)

incarceration_long['race'] = incarceration_long['race'].str.title()
display(incarceration_long.head())
```


## 5. Visualization: Incarceration Trends Over Time
This chart compares incarceration rates for Black and White populations from 2011 to 2021.


```python

plt.figure(figsize=(10, 6))

for race in incarceration_long['race'].unique():
    subset = incarceration_long[incarceration_long['race'] == race]
    plt.plot(subset['year'], subset['incarceration_rate'], marker='o', label=race)

plt.title('Incarceration Rates by Race (2011–2021)')
plt.xlabel('Year')
plt.ylabel('Incarceration Rate')
plt.legend()
plt.tight_layout()
plt.savefig(output_dir / 'incarceration_trends_by_race.png', dpi=300, bbox_inches='tight')
plt.show()
```


## 6. Measuring the Gap
This section calculates the incarceration-rate gap between Black and White populations each year.


```python

incarceration['gap_black_minus_white'] = incarceration['black'] - incarceration['white']
display(incarceration[['year', 'white', 'black', 'gap_black_minus_white']])
```

```python

plt.figure(figsize=(10, 6))
plt.plot(incarceration['year'], incarceration['gap_black_minus_white'], marker='o')
plt.title('Gap in Incarceration Rates: Black Minus White (2011–2021)')
plt.xlabel('Year')
plt.ylabel('Rate Difference')
plt.tight_layout()
plt.savefig(output_dir / 'incarceration_gap.png', dpi=300, bbox_inches='tight')
plt.show()
```


## 7. Visualization: Disenfranchisement Rates by Race/Ethnicity (2022)
This chart compares 2022 disenfranchisement rates across groups.


```python

plt.figure(figsize=(10, 6))
plt.bar(disenfranchisement['category'], disenfranchisement['value_numeric'])
plt.title('Disenfranchisement Rates by Race/Ethnicity (2022)')
plt.xlabel('Group')
plt.ylabel('Rate (%)')
plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig(output_dir / 'disenfranchisement_rates_2022.png', dpi=300, bbox_inches='tight')
plt.show()
```


## 8. Key Findings


```python

avg_white = incarceration['white'].mean()
avg_black = incarceration['black'].mean()
avg_gap = incarceration['gap_black_minus_white'].mean()

highest_group = disenfranchisement.loc[disenfranchisement['value_numeric'].idxmax(), 'category']
highest_rate = disenfranchisement['value_numeric'].max()

print(f"Average incarceration rate for White population (2011–2021): {avg_white:.2f}")
print(f"Average incarceration rate for Black population (2011–2021): {avg_black:.2f}")
print(f"Average Black-White incarceration gap: {avg_gap:.2f}")
print(f"Highest disenfranchisement rate in 2022: {highest_group} ({highest_rate:.2f}%)")
```


## 9. Interpretation

### Main Takeaways
- In every year from **2011 to 2021**, the incarceration rate for the Black population was far higher than the rate for the White population.
- Although incarceration rates declined for both groups over time, the disparity remained substantial throughout the full period.
- In the 2022 disenfranchisement dataset, **African American** people had the highest reported disenfranchisement rate at **4.97%**, which was notably higher than the rates for Latinx, White, and Overall categories.
- Taken together, these datasets suggest that racial inequality appears not only in incarceration outcomes, but also in downstream civic consequences such as reduced voting access.

### Why This Matters
This analysis shows how data can help reveal structural inequality. Even with a small dataset, the pattern is clear: criminal justice outcomes and voting access are not experienced equally across groups.

The findings show persistent racial disparities in incarceration rates and unequal disenfranchisement outcomes. These patterns connect numerical analysis to broader real-world issues involving justice, representation, and civic participation.
