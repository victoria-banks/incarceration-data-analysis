# Data Analysis of Incarceration & Disenfranchisement


```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
inc = pd.read_csv('../data/incarceration.csv')
dis = pd.read_csv('../data/disenfranchisement_data.csv')
```

```python
inc.columns = inc.columns.str.lower()
dis.columns = dis.columns.str.lower()
dis['value'] = dis['value'].str.replace('%','').astype(float)
```

```python
inc_long = inc.melt(id_vars='year', value_vars=['white','black'], var_name='race', value_name='rate')
```

```python
for r in inc_long['race'].unique():
    d = inc_long[inc_long['race']==r]
    plt.plot(d['year'], d['rate'], label=r)
plt.legend()
plt.title('Incarceration Trends by Race')
plt.xlabel('Year')
plt.ylabel('Rate')
plt.show()
```

```python
inc['gap'] = inc['black'] - inc['white']
plt.plot(inc['year'], inc['gap'])
plt.title('Black-White Incarceration Gap')
plt.show()
```

```python
plt.bar(dis['category'], dis['value'])
plt.title('Disenfranchisement Rates 2022')
plt.show()
```
